```python
import random
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python

avg=10
cofficient=0.5
print(cofficient)


errors=[1,4,-1,0,-3]

# for i in range(0,5):
#   errors=[]
#   er=random.randint(-3,4)
#   errors.append(er)
#   print(errors)
predicted=[10]
left=[]
for i in range(0,len(errors)):
  
  
  if i==0:
    le=predicted[i]+errors[i]
    left.append(le)
    print("left list",left)
  else:
      new_predicted=avg+cofficient*errors[i-1]
      predicted.append(new_predicted)
      le1=predicted[i]+errors[i]
      left.append(le1)
print("final_last",left)

print("predicted_list:",predicted)

    
```

```python
rows=[]
for step,err,pred,lf in zip(range(len(errors)),errors,predicted,left):
   rows.append([step, err, pred, lf])
print(tabulate(rows, headers=["Step","Error","Prediction","Left"],tablefmt="grid"))
```

```python
#plot graph
t_values = list(range(1, len(errors)+1))
plt.plot(t_values, [avg]*len(errors), label="Average (U)", linestyle="--", color="red")
plt.plot(t_values, predicted, label="J values", marker="o", color="blue")
plt.xlabel("Time Period (t)")
plt.ylabel("Values")
plt.title("Average (U) vs J values")
plt.legend()
plt.grid(True)
plt.show()

plt.xlabel("avg")
plt.xlabel("Value")
```
