```python
import numpy as np
from collections import Counter
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

def dist(a, b):
    return np.sqrt(np.sum((a - b)**2))

def get_label(xtr, ytr, p, k):
    d = [dist(p, x) for x in xtr]
    idx = np.argsort(d)[:k]
    labs = [ytr[i] for i in idx]
    return Counter(labs).most_common(1)[0][0]

def run_knn(xtr, ytr, xts, k):
    out = []
    for p in xts:
        out.append(get_label(xtr, ytr, p, k))
    return np.array(out)

data = load_iris()
X = data.data
y = data.target

xtr, xts, ytr, yts = train_test_split(X, y, test_size=0.25, random_state=1)

k = 3
pred = run_knn(xtr, ytr, xts, k)

print("Accuracy:", accuracy_score(yts, pred))
print("Pred:", pred[:10])
print("True:", yts[:10])
```
