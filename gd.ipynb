```python
import numpy as np

def gd(x, y, lr=0.01, it=500):
    w = 0
    b = 0
    n = len(x)
    for _ in range(it):
        yp = w*x + b
        dw = (-2/n)*sum(x*(y-yp))
        db = (-2/n)*sum(y-yp)
        w = w - lr*dw
        b = b - lr*db
    return w, b

x = np.array([1,2,3,4,5], dtype=float)
y = np.array([2,4,5,4,5], dtype=float)

w, b = gd(x, y)
print("w:", w, "b:", b)
print("pred:", w*x + b)
```
