```python
from tabulate import tabulate
import matplotlib.pyplot as plt
```

```python
from tabulate import tabulate


mean = 50
coefficient = 0.5
actual = [50, 49, 52, 48, 45.5]  


errors = []
forecast_value = []


for i in range(len(actual)):
    if i == 0:  
        forecast = mean
    else:      
        forecast = mean + coefficient * errors[i-1]
    
    forecast_value.append(forecast)

  
    err = actual[i] - forecast
    errors.append(err)

table = []
for i in range(len(actual)):
    table.append([i, actual[i], round(forecast_value[i], 2), round(errors[i], 2)])

print(tabulate(table, headers=["Step", "Actual", "Forecast", "Error"], tablefmt="grid"))

```

```python
# Plot
plt.figure(figsize=(8,5))
plt.plot(range(len(actual)), actual, marker='o', label="Actual")
plt.plot(range(len(forecast_value)), forecast_value, marker='s', linestyle="--", label="Forecast")
plt.axhline(y=mean, color='r', linestyle=':', label="Mean (50)")

plt.title("Actual vs Forecasted Values (MA1 Model)")
plt.xlabel("Step")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()

```
